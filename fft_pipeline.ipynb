```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

# --- Simulated vibration signal (you can replace with real data later) ---
fs = 1000  # sampling frequency (Hz)
t = np.linspace(0, 1, fs)  # 1 second of data

# vibration = 50 Hz + noise
signal = 0.8*np.sin(2*np.pi*50*t) + 0.2*np.random.randn(len(t))

# --- FFT ---
fft_vals = np.abs(rfft(signal))
freqs = rfftfreq(len(signal), 1/fs)

# --- Plot ---
plt.plot(freqs, fft_vals)
plt.title("FFT of Vibration Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
import os
import pandas as pd
import numpy as np
from scipy.stats import kurtosis
import matplotlib.pyplot as plt

# 1. SETUP PATHS
# Make sure this matches your folder name exactly
data_dir = "data/1st_test/" 

# Get all files and sort them so they are in chronological order
files = sorted([f for f in os.listdir(data_dir) if os.path.isfile(os.path.join(data_dir, f))])

# 2. DATA CONTAINERS
kurt_results = []
rms_results = []
timestamps = []

print(f"Starting analysis on {len(files)} files...")

# 3. THE ENGINEERING LOOP
# We process every 10th file ([::10]) so it doesn't take 20 minutes to run
for filename in files[::10]:
    try:
        file_path = os.path.join(data_dir, filename)
        # NASA Test 1 uses Tab separation
        df = pd.read_csv(file_path, sep='\t', header=None)
        
        # We focus on Bearing 3 (Column 4) because it's the one that crashes
        signal = df[4].values
        
        # Calculate our 'Condition Indicators'
        rms = np.sqrt(np.mean(signal**2))
        kurt = kurtosis(signal)
        
        kurt_results.append(kurt)
        rms_results.append(rms)
        timestamps.append(filename) # Keep track of the time
        
    except Exception as e:
        print(f"Skipping {filename} due to error: {e}")

print("Analysis complete! Plotting results...")

# 4. VISUALIZE THE FAILURE
plt.figure(figsize=(12, 6))
plt.plot(kurt_results, label="Kurtosis (Bearing 3)", color='red')
plt.axhline(y=3, color='black', linestyle='--', label="Healthy Baseline")
plt.title("The 'Death Curve': Bearing Degradation Over Time")
plt.ylabel("Value")
plt.xlabel("Time (Every 10th File)")
plt.legend()
plt.grid(True)
plt.show()
```
