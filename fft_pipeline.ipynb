```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq

# --- Simulated vibration signal (you can replace with real data later) ---
fs = 1000  # sampling frequency (Hz)
t = np.linspace(0, 1, fs)  # 1 second of data

# vibration = 50 Hz + noise
signal = 0.8*np.sin(2*np.pi*50*t) + 0.2*np.random.randn(len(t))

# --- FFT ---
fft_vals = np.abs(rfft(signal))
freqs = rfftfreq(len(signal), 1/fs)

# --- Plot ---
plt.plot(freqs, fft_vals)
plt.title("FFT of Vibration Signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import kurtosis

rms = np.sqrt(np.mean(signal**2))
peak_freq = freqs[np.argmax(fft_vals)]
kurt = kurtosis(signal)

print("RMS:", rms)
print("Peak frequency:", peak_freq)
print("Kurtosis:", kurt)
```
