```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup Data
fs = 1000  
t = np.linspace(0, 1, fs)
# random signal
signal = np.sin(2 * np.pi * 50 * t) + 0.5 * np.random.randn(len(t))

# 2. Calculate FFT
fft_values = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(t), 1/fs)

# 3. Create the 2-Graph Plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
plt.subplots_adjust(hspace=0.4)

# Top Plot: The messy raw data
ax1.plot(t[:100], signal[:100]) 
ax1.set_title("Time Domain: Raw Vibration")
ax1.set_xlabel("Time (seconds)")
ax1.set_ylabel("Amplitude")

# Bottom Plot: The clean frequency spike
ax2.plot(frequencies[:fs//2], np.abs(fft_values)[:fs//2], color='red')
ax2.set_title("Frequency Domain: FFT Result")
ax2.set_xlabel("Frequency (Hz)")
ax2.set_ylabel("Intensity")

plt.show()
```
